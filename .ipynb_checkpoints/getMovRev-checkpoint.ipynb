### 데이터 크롤링
* url= https://movie.naver.com/movie/point/af/list.nhn?&page=1

```python
#html 페이지 가져오기 라이브러리
import requests
from bs4 import BeautifulSoup
```

#### http response 결과값
* 200 정상회신
* 404 요청 페이지 없음
* 500 내부에러

```python
def myCrawl(t):
    #print(t.get_text())
    #고유번호
    no=t.select('.ac')[0].get_text()

    #영화제목
    title=t.select('.color_b')[0].get_text()

    #별점
    point=t.select('em')[0].get_text()

    #리뷰내용
    src=str(t.select('.title')[0])
    startPos=src.find('<br/>')
    endPos=src.find('<a class="report"')
    content=src[startPos+5:endPos].strip()

    #날짜
    mdate=str(t.select('.num')[1])
    msrc='20'+mdate[-13:-5]
    mday=msrc.replace('.','')

    #print(no,title,point,content,mday)
    #print('*'*50)

    return (no,title,point,content,mday)
```

```python
def myReview():
    result=[]
    urlsrc='https://movie.naver.com/movie/point/af/list.nhn?&page={0}'
    for i in range(1,10):
        url=urlsrc.format(i)
        print(url)
        res=requests.get(url)
        html=res.text.strip()    #html요소를 doc object형태로 추출 
        soup=BeautifulSoup(html,'html.parser')     #css 셀렉터를 활용하여 타겟 데이터 추출   #타겟은 페이지에서 유일한 요소여야하며 획득하고자 하는 데이터를 모두 포함하여야한다.
        tgt=soup.select('table.list_netizen tbody tr') #array 형태로 나옴   #len(tgt) #1이여야함   # tgt[0]
        for t in tgt:
            res=myCrawl(t)
            result.append(res)
    return result
```

```python
resList=myReview()
resList
```

```python
#데이터 정제
resList[:10]
cnt=0
pnts=0
for r in resList:
    if(r[1]=='정직한 후보'):
        cnt+=1
        pnts+=int(r[2])
print('평점:',round(pnts/cnt,2),'총 평수:',cnt)
```

```python
#데이터 정제
resList[:10]
movies={}
for r in resList:
    key=r[1]
    if(key in movies.keys()):
        movies[key]+=int(r[2])
    else:
        movies[key]=int(r[2])
movies
```
