### 데이터 크롤링
* url=https://movie.naver.com/movie/point/af/list.nhn?&page=1

```python
# html 페이지 가져오기 라이브러리
import requests
from bs4 import BeautifulSoup 
import sqlite3
```

#### http response 결과값
* 200 정상회신
* 404 요청 페이지 없음
* 500 내부에러

```python
def myCrawl(t):
    nos=t.select('.ac')
    no=nos[0].get_text()
    title=t.select('.color_b')[0].get_text()
    point=t.select('em')[0].get_text()
    
    src=str(t.select('.title')[0])
    startPos=src.find('<br/>')
    endPos=src.find('<a class="report"')
    content=src[startPos+5:endPos].strip()

    mdate=t.select('.num')
    msrc='20'+str(mdate[1])[-13:-5]
    mday=msrc.replace('.','')
    return (no,title,point,content,mday)
```

```python
def myReview(frpage=1,topage=100):
    result=[]
    urlsrc='https://movie.naver.com/movie/point/af/list.nhn?&page={0}'
    for i in range(frpage,topage):
        url=urlsrc.format(i)
        print(url)
        res=requests.get(url)
        html =res.text.strip()
        soup=BeautifulSoup(html,'html.parser') 
        tgt=soup.select('table.list_netizen tbody')
        tgts=tgt[0].select('tr')
        for t in tgts:
            res=myCrawl(t)
            result.append(res)
    return result
```

```python
def dbPrs(data):
    dbname='movie.db'
    data=list(data)
    checkid=data[0]
    print("checkid:",checkid)
    with sqlite3.connect(dbname) as conn:
        cur=conn.cursor()
        sql='select idx from reviews where idx={0}'.format(checkid)
        res=cur.execute(sql)
        resultid=res.fetchall()
        if(resultid):
            print("중복아이디가 있습니다.")
        else:
            data[1]=data[1].replace("'","")
            data[3]=data[3].replace("'","")
            if(data[3]==''):
                data[3]=' '
            tmp="insert into reviews (idx,title,content,point,rday) values({0},'{1}','{3}',{2},{4})"
            sql=tmp.format(data[0],data[1],data[2],data[3],data[4])
            print(sql)
            cur.execute(sql)
            conn.commit()
            
            print("데이터 입력:"+data[1])
        cur.close()
```

```python
resList=myReview(1,100)
for res in resList:
    dbPrs(res)
```
